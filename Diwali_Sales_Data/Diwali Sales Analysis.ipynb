# **Exploratory Data Analysis**

# ***Gender***


```python
df.columns
```

```python
ax = sns.countplot(x = 'Gender', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# values are not displayed over the barr
# for this solution is given below



```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by="Amount", ascending=False)
```

```python
sales_gem = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by="Amount", ascending=False)
```

```python
sales_gem
```

```python
sns.barplot(data=sales_gem, x='Gender', y='Amount')
```

# ***Age***

```python
sales_gem = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by="Amount", ascending=False)
ax = sns.barplot(y='Amount', x='Age Group', data=sales_gem)
```

```python
ax = sns.countplot(data=df, x='Age Group', hue='Gender')
for bar in ax.containers:
    ax.bar_label(bar)
```
