### ***Number of Mathes Played by RCB and WON***

```python
mask1 = mathes['team1'] == 'Royal Challengers Bangalore'
mask2 = mathes['team2'] == 'Royal Challengers Bangalore'
mask3 = mathes['winner'] == 'Royal Challengers Bangalore'
print('Matches : ',mathes[(mask1 | mask2) ].shape[0])
print('Matches WON : ',mathes[(mask1 | mask2) & mask3].shape[0])
print('Winning percentage : ', (mathes[(mask1 | mask2) & mask3].shape[0]/mathes[(mask1 | mask2) ].shape[0])*100)
```

## ***Return the list of Total Matches played by every team***

### Method 1

```python
def total_matches_played_by(team):
    mask1 = mathes['team1'] == team
    mask2 = mathes['team2'] == team
    return mathes[mask1 | mask2].shape[0]
```

```python
list_of_teams
```

```python
played = []
for team in list_of_teams:
    played.append(total_matches_played_by(team))
played
```

```python
new_df = pd.DataFrame({
    'Team': list_of_teams,
    'Matches Played': played
})
new_df
```
